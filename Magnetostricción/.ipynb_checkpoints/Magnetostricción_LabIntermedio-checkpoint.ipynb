```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import pandas as pd
import uncertainties as unc

def linear_function(X, a, b):
    return a * X + b
```

```python
A1 = np.array([-3.15,-3,-2.5,-2,-1.5,-1,-0.5,0.5,1,1.5,2,2.5,3,3.25])
B1 = np.array([-59,-55.6,-46.6,-37.7,-27.8,-18.6,-9.6,9.5,18.6,28.2,37.4,46.3,55.8,60.3])
sigma_B1 = np.array([0.1]*len(B1))
sigma_A1 = np.array([0.01]*len(A1))
```

```python
popt1, pcov1 = curve_fit(linear_function, A1, B1, sigma=sigma_B1, absolute_sigma=True)
```

```python
slope = popt1[0]
intercept = popt1[1]

slope_std_err = np.sqrt(pcov1[0, 0])
intercept_std_err = np.sqrt(pcov1[1, 1])

Y_pred = linear_function(A1, *popt1)


residuals = (B1 - Y_pred)/sigma_B1

print(f"Slope: {slope} ± {slope_std_err}")
print(f"Intercept: {intercept} ± {intercept_std_err}")
print(f"Residuals: {residuals}")

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10), gridspec_kw={'height_ratios': [2, 1]})  # 2 rows, 1 column, width = 10 inches, height = 10 inches

ax1.errorbar(A1, B1, yerr= sigma_B1,xerr=sigma_A1,  fmt='o', label='Datos', color="black")
ax1.plot(A1, Y_pred, color='black', linewidth=1, label=f'Ajuste lineal: B($I$) = ({slope:.2f}± {slope_std_err:.2f})$I$ + ({intercept:.2f}± {intercept_std_err:.2f})')

ax1.set_xlabel('I(A)')
ax1.set_ylabel('B(mT)')
ax1.legend(loc='upper left')
ax1.set_title('Caracterización de la Bobina')

# Plot normalized residuals
ax2.scatter(A1, residuals, color='black', alpha=1, label='Residuales')
ax2.axhline(0, color='black', linewidth=1, linestyle='--')
ax2.set_xlabel('I(A)')
ax2.set_ylabel('Residuales Normalizados')
ax2.legend(loc='upper right')
ax2.set_title("Residuales Normalizados de la Regresión Lineal")

plt.tight_layout()

# Save the figure as a PDF file
plt.savefig('Weighted_linear_regression_with_normalized_residuals_curve_fit.pdf', format='pdf')
plt.show()
```

```python
A_zinc = np.array([-3.4,-1.72,-1.1,-0.77,-0.58,-0.23,0,0.29,0.52,0.8,1.04,1.7,3.01])
B_pred1_unc  = []

for i in range(len(A_zinc)):
  B_unc_i = unc.ufloat(slope,slope_std_err)*unc.ufloat(A_zinc[i],sigma_A1[i]) + (unc.ufloat(intercept,intercept_std_err))
  B_pred1_unc.append(B_unc_i)

B_pred1 = []
B_pred1_sigma = []

for i in B_pred1_unc:
  B_pred1.append(i.nominal_value)
  B_pred1_sigma.append(i.std_dev)

B_pred1 = np.array(B_pred1)
B_pred1_sigma = np.array(B_pred1_sigma)
B_pred1
```

```python
plt.errorbar(B_pred1,Min1,fmt="o--",xerr = B_pred1_sigma,color="black")
plt.ylabel("$|\Delta L|$ [nm]")
plt.xlabel("Campo magnético [mT]")
plt.title("Magnetostricción Níquel")
```
